```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
```

```python
def table_to_df(URL):
    html = requests.get(URL)
    soup = BeautifulSoup(html.text)
    tables = soup.find_all("table")
    output = []
    for table in tables:
        output_rows = []
        for table_row in table.findAll('tr'):
            columns = table_row.findAll('td')
            output_row = []
            for column in columns:
                output_row.append(column.text)
            output_rows.append(output_row)
        output.append(output_rows)
    return output
```

```python
URL = 'https://www.mohfw.gov.in/'
df = pd.DataFrame(table_to_df(URL)[1])
df = df.drop(0,axis=1)
df = df.drop(0,axis=0)
```

```python
df.columns = ['States', 'Total Confirmed cases (Indian National)','Total Confirmed cases ( Foreign National )','Cured/Discharged/Migrated','Death'] 
```

```python
ind = []
for i in range(len(df)):
    ind.append(i)
df.index = ind
df
```
